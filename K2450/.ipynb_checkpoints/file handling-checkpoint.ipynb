```python
import csv
```

```python
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randint
%matplotlib inline 
```

```python
def csv_init():
    with open('ECR.csv', 'w', encoding = 'UTF8', newline = '') as f:
        writer = csv.writer(f)
        writer.writerow(["Time", "Bias", "Current"])
    f.close()
```

```python
def csv_write(dataRow):
    with open('ECR.csv', 'a', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(dataRow)
    f.close()
```

```python
with open('ECR.csv') as f:
    lines = f.read().splitlines()
f.close()
```

```python
exp_time, exp_bias, exp_curr = [], [], []

for data in open('ECR.csv', 'r'):
    time = float(data.split(",")[0])
    bias = float(data.split(",")[1])
    current = float(data.split(",")[2].strip('""\n'))
    
    exp_time.append(time)
    exp_bias.append(bias)
    exp_curr.append(current)
    
    print(time, bias, current)
```

```python
# datetime object containing current date and time
now = datetime.now()
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")

plt.plot(exp_time,exp_bias)
plt.plot(exp_time,exp_curr)
plt.xlabel('Applied bias (V)')
plt.ylabel('Measured current (A)')
plt.title('Electrochemical nanopore generation experiment (' + dt_string + ')')
```

```python
# datetime object containing current date and time
now = datetime.now()
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")

fig, axes = plt.subplots(figsize=(9,6))
bias_ax = axes.twinx()
axes.plot(exp_time, exp_bias)
bias_ax.plot(exp_time, exp_curr)
```
