```python
import csv
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
ex_time, ex_bias, ex_curr = [], [], []

filename = "ECR_FIBhole_1sec_0_1V_"
extension = "csv"
index = 3

filename_str = filename + str(index) + "." + extension

with open(filename_str, "r", encoding = "UTF-8") as file:
    reader = csv.reader(file)
    for lines in reader:
        bTime = float(lines[0])
        bBias = float(lines[1])
        bCurr = float(lines[2])

        ex_time.append(bTime)
        ex_bias.append(bBias)
        ex_curr.append(bCurr)
file.close()

plt.plot(ex_time, ex_curr)
```
